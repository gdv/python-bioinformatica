Rivediamo l'esercizio 1.2


````python
import re
nuovo_testo = re.sub('\<ref\>(.*?)\<\/ref\>', '', testo)
nuovo_testo = re.sub('[\.,;:\"\?]+($|\s+)', ' ', nuovo_testo)
nuovo_testo = re.sub('(^|\s+)[\"\+]+', ' ', nuovo_testo)
nuovo_testo = re.sub('(^|\s+)[\'\{\[\(=\"\']+', ' ', nuovo_testo)
nuovo_testo = re.sub('[\}\]\)=\"\']+($|\s+)', ' ', nuovo_testo)
nuovo_testo = re.sub('[\|\n]', ' ', nuovo_testo)
nuovo_testo = re.sub(' \s+', ' ', nuovo_testo)

print(nuovo_testo)
````

````python
parole = re.split(' ', nuovo_testo)
````

````python
from collections import Counter
conteggio = Counter()
for p in parole:
    conteggio[p] += 1
````

````python
conteggio
````

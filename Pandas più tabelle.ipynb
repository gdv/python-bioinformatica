Il campo `gene` contiene la stringa `gene:` che  inutile: rimuovere questa stringa.

```python
def togli(str):
    return str.strip('gene:')

names['gene_pulito'] = names['gene'].apply(togli)
```

Andiamo a pulire la colonna gene_symbol dai dati che dovrebbero essere assegnati a description.

Il primo passo  individuare le righe da pulire.

```python
missing_desc = names['description'].isnull()
missing_desc
```

*Modifichiamo* i dati una colonna alla volta.
Usiamo una forma di fancy indexing sfruttando `missing_desc` per le righe. Siccome stiamo lavorando su una selezione di righe (slice), pandas richiede si acceda tramite `loc`.

```python
names.loc[missing_desc, 'description'] = names[missing_desc]['gene_symbol']
names.loc[missing_desc, 'gene_symbol'] = None
names.head()
```

```python
names.head()
```

Adesso si possono pulire le colonne, esattamente come abbiamo fatto per `gene`

L'idea  fare un ciclo su tutte le colonne da pulire e, per ogni colonna:

1. scrivere una funzione di nome `pulisci` che pulisce quella singola colonna
2. usare `apply` per applicare la funzione `pulisci`

Notare che la funzione `pulisci` viene ridefinita per ogni colonna. In questo modo si pu usare una `apply` che richiede una funzione con un singolo argomento.

```python

for colonna in ['chromosome', 'gene', 'gene_biotype', 'transcript_biotype', 'gene_symbol', 'description']:
    def pulisci(campo):
        if campo != None:
            return re.sub(colonna + ':', '', campo)
        else:
            return None
        
    names[colonna + '_pulita'] = names[colonna].apply(pulisci)
names.head()
```
